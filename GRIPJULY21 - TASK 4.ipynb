# Renaming columns

```python
data.rename(columns={'iyear':'Year','imonth':'Month','iday':"day",'gname':'Group','country_txt':'Country',
                     'region_txt':'Region','provstate':'State','city':'City','latitude':'latitude','longitude':'longitude',
                     'summary':'summary','attacktype1_txt':'Attacktype','targtype1_txt':'Targettype',
                     'targsubtype1_txt':'Targetsubtype','weaptype1_txt':'Weapon','nkill':'kill','nwound':'Wound'},inplace=True)
```

```python
data.head()
```

By seeing the above data we can say that there are many columns with same name so first of all I'm just selecting some particular features for my Exploratory Data Analysis

```python
new_data= data[['Year','Month','day','Country','State','City','Region',"Attacktype",
                "Targettype",'Targetsubtype','kill', 'Wound','target1','summary','Group','Weapon','motive']]

new_data.head()
```

```python
new_data.shape
```

```python
new_data.dtypes
```

# Understanding the Missing Value data

```python
new_data.isnull().sum()/len(new_data)*100
```

```python
new_data.drop(['summary','motive'],axis=1,inplace=True)
```

```python
new_data
```

```python
new_data.isnull().sum()
```

```python
new_data['State'].fillna(value="Not specified",inplace=True)
new_data['City'].fillna(value="Not specified",inplace=True)
new_data['target1'].fillna(value="Not known",inplace=True)
new_data['Targetsubtype'].fillna(value="Not Specified",inplace=True)
```

```python
print(new_data['kill'].mode())
print(new_data['Wound'].mode())
```

```python
# Now filling the misisng value data with mode
new_data['Wound'] = new_data['Wound'].fillna(0)
new_data['kill'] = new_data['kill'].fillna(0)
```

```python
new_data.isnull().sum()
```

```python
new_data.shape
```

```python
new_data.head()
```

```python
plt.figure(figsize=(10,7))
sns.heatmap(new_data.corr(),annot=True)
```

```python
new_data['Year'].value_counts().sort_index()
```

```python
year = new_data['Year'].unique()
count_of_years = new_data['Year'].value_counts().sort_index()
plt.figure(figsize = (15,8))
sns.barplot(x = year,y = count_of_years,palette = 'gist_earth_r')
plt.xticks(rotation = 55)
plt.xlabel(' Year of Attack ')
plt.ylabel(' Number of Attacks each year ')
plt.title(' Attacks_per_Year ')
plt.show()
```

So from the above visualization wwe can say that in 2014, there are maximum number of attacks i.e. 16794.

```python
# Region wise casualities i.e. how many peoples are killed and how many are in critical condition.
```

```python
new_data.groupby(['Region']).count()['kill']
```

```python
plt.subplots(figsize=(12,7))
new_data.groupby(['Region']).count()['kill'].plot()
plt.xticks(rotation=90)
plt.ylabel("Casualities")
```

```python
new_data.groupby(['Region']).count()['Wound']
```

```python
plt.subplots(figsize=(12,8))
new_data.groupby(['Region']).count()['Wound'].plot()
plt.xticks(rotation=90)
plt.ylabel("Casualities")
```

```python
# Attack type
```

```python
new_data['Attacktype'].value_counts().plot(kind='bar',figsize=(12,7),color='cyan')
plt.xticks(rotation = 70)
plt.xlabel("Attack-Type",fontsize=15)
plt.ylabel("Number of Attacks",fontsize=15)
plt.title("Type of  Attack",fontsize=20)
plt.show()
print("Maximum type of Attack are Bombing\Explosions.")
```

```python
# Most used weapon for the Attack.
```

```python
plt.figure(figsize= (12,8))
sns.barplot(x = new_data.Weapon.value_counts()[:6].index , y = new_data.Weapon.value_counts()[:6].values)
plt.title('Type of Weapon most used for the Attack')
plt.xlabel('Weapon Type')
plt.ylabel('Count')
plt.xticks(rotation = 90)
plt.show()
print(" So we can say that 'Explosives' are mostly used for the Attack.")
```

```python
# Number of attacks in a particular Region,Country,State,City.
```

```python
plt.subplots(figsize=(12,6))
sns.countplot('Region',data=new_data,palette='YlGnBu_r',order=new_data['Region'].value_counts().index)
plt.xticks(rotation=75)
plt.xlabel('Region',fontsize=14)
plt.ylabel("Number of Attacks",fontsize=14)
plt.title('Number Of Attacks By Region')
plt.show()
print("We can conclude that the Region which was mostly attacked is 'Middle East and North Africa'. ")
```

```python
plt.subplots(figsize=(14,6))
sns.countplot('Country',data=new_data,palette='PiYG',order=new_data['Country'].value_counts()[:35].index)
plt.xticks(rotation=90)
plt.xlabel('Country',fontsize=14)
plt.ylabel("Number of Attacks",fontsize=14)
plt.title('Number Of Attacks By Country')
plt.show()
print("We can conclude that the Country which was mostly attacked is 'Iraq'. ")
```

```python
plt.subplots(figsize=(12,6))
sns.countplot('State',data=new_data,palette='Paired',order=new_data['State'].value_counts()[:20].index)
plt.xticks(rotation=75)
plt.xlabel('State',fontsize=15)
plt.ylabel("Number of Attacks",fontsize=14)
plt.title('Number Of Attacks By State')
plt.show()
print("We can conclude that the State which was mostly attacked is 'Baghdad'. ")
```

```python
plt.subplots(figsize=(12,6))
sns.countplot('City',data=new_data,palette='tab20c',order=new_data['City'].value_counts().drop('Unknown')[:20].index)
plt.xticks(rotation=75)
plt.xlabel('City',fontsize=15)
plt.ylabel("Number of Attacks",fontsize=14)
plt.title('Number Of Attacks in a particular City',fontsize=15)
plt.show()
print("We can conclude that the City which was mostly attacked is 'Baghdad'. ")
```

```python
new_data[['Attacktype','kill']].groupby(["Attacktype"],axis=0).sum().plot(kind='bar',figsize=(10,8),color=['yellow'])
plt.xticks(rotation=70)
plt.title(" Number of peoples killed in which Attack ",fontsize=20)
plt.ylabel(' Number of peoples killed ',fontsize=15)
plt.xlabel(' Type of Attack ',fontsize=15)
plt.show()
print("Most of the peoples are killed by 'Armed Assault' followed by Bombing/Explosion .")
```

```python
new_data['Group'].value_counts().to_frame().drop('Unknown').head(10).plot(kind='bar',color='green',figsize=(16,11))
plt.xticks(rotation=78)
plt.title("Top 10 terrorist group attack",fontsize=20)
plt.xlabel("Terrorist Group name",fontsize=15)
plt.ylabel("Attack number",fontsize=15)
plt.show()
```

```python
Target_data = new_data['Targettype'].value_counts().nlargest(n=12)
Target_data
```

```python
plt.figure(figsize= (16,10))
sns.barplot(x = new_data['Targettype'].value_counts().nlargest(n=12).index , y = new_data['Targettype'].value_counts().nlargest(n=12).values,palette='copper_r')
plt.title('Top 12 Target Type',fontsize=15)
plt.xlabel('Target Type',fontsize=15)
plt.ylabel('Attacks count',fontsize=15)
plt.xticks(rotation = 90)
plt.show()
print('Top target type is "Private citizens and Property".')
```

```python
plt.figure(figsize= (16,10))
sns.barplot(x = new_data['target1'].value_counts().drop('Unknown').nlargest(n=15).index , y = new_data['target1'].value_counts().drop('Unknown').nlargest(n=15).values,palette='copper_r')
plt.title('Top 15 Targets',fontsize=15)
plt.xlabel('Target Type',fontsize=15)
plt.ylabel('Attacks count',fontsize=15)
plt.xticks(rotation = 90)
plt.show()
print("Top Target's are innocent 'Civilians' followed by Soldiers.")
```
